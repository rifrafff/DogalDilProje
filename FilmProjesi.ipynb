```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

def remove_stopwords(text):
    if isinstance(text, str):
        stop_words = set(stopwords.words('english'))
        words = word_tokenize(text)
        filtered_words = [word for word in words if word.lower() not in stop_words]
        return ' '.join(filtered_words)
    else:
        return text
```

```python
new_column = 'temizlenmis_konu'
df[new_column] = df[target_column].apply(remove_stopwords)
```

```python
from simplemma import text_lemmatizer

def clean_text(text):
    if isinstance(text, str):
        text = text.replace("Â", "a")
        text = text.replace("â", "a")
        text = text.replace("î", "i")
        text = text.replace("Î", "ı")
        text = text.replace("İ", "i")
        text = text.replace("I", "ı")
        text = text.replace(u"\u00A0", " ")
        text = text.replace("|", " ")

        text = re.sub(r"@[A-Za-z0-9]+", " ", text)
        text = re.sub(r"(.)\1+", r"\1\1", text)
        text = re.sub(r"https?:\/\/\S+", " ", text)
        text = re.sub(r"http?:\/\/\S+", " ", text)
        text = re.sub(r"\n", " ", text)
        text = re.sub(r"#(\w+)", " ", text)
        text = re.sub(r"^\x00-\x7F]+", " ", text)
        text = re.sub(r"[^A-Za-zâîığüşöçİĞÜŞÖÇ]+", " ", text)
        text = re.sub(r"((https://[^\s]+))", " ", text)
        text = " ".join(text.lower().strip().split())

        return " ".join(text_lemmatizer(text, lang="tr"))
    else:
        return text
```

```python
def clean_row(row):
    konu = clean_text(row["temizlenmis_konu"])
    return konu
```

```python
df["clean"] = df.apply(lambda row: clean_text(row["temizlenmis_konu"]), axis=1)
df.head()
```

```python
# Kategorileri belirleyin (örneğin: "aksiyon", "komedi", "drama" vb.)
categories = df["Türler"].unique()

# Hangi film hangi kategoriye ait olduğunu etiketleyin
def etiketle(row):
    for key, category in enumerate(categories):
        if row["Türler"] == category:
            return key

df["labels"] = df.apply(lambda row: etiketle(row), axis=1)
```

```python
def build_markov_model(corpus, order=1):
    markov_model = defaultdict(list)

    # Corpus üzerinde dolaşılarak Markov modeli oluşturuluyor
    for i in range(len(corpus) - order):
        current_state = tuple(corpus[i:i + order])
        next_state = corpus[i + order]
        markov_model[current_state].append(next_state)

    return markov_model
```

```python
def generate_text(markov_model, seed, length=10):
    current_state = tuple(seed.split())
    generated_text = list(current_state)

    for _ in range(length):
        if current_state in markov_model:
            next_state = random.choice(markov_model[current_state])
            generated_text.append(next_state)
            current_state = tuple(generated_text[-len(current_state):])
        else:
            break

    return ' '.join(generated_text)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, f1_score, classification_report
import numpy as np

# Veri setini yükle
df = pd.read_excel("film_Datasi.xlsx")

# NaN değerleri temizle
df = df.dropna(subset=['Konu'])  # 'konu' sütunu içindeki NaN değerleri temizle

# 'konu' sütunu yerine veri setinizdeki metin sütununun gerçek adını kullanın
metin_sutunu_adi = 'Konu'  # Eğer farklı bir isim kullanıyorsanız, burada düzenleme yapın

tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(df[metin_sutunu_adi])

# Veriyi eğitim ve test setine ayır
X_train, X_test, y_train, y_test = train_test_split(tfidf_matrix, df['Türler'], test_size=0.2, random_state=42)

# Destek Vektör Makineleri (SVM) modelini tanımla ve eğit
model_SVM = SVC(random_state=42)
model_SVM.fit(X_train, y_train)

# Random Forest modelini tanımla ve eğit
model_RF = RandomForestClassifier(random_state=42)
model_RF.fit(X_train, y_train)

# Kullanıcı girdisi için TF-IDF vektörü oluştur
kullanici_konu_vectorized = tfidf_vectorizer.transform(["Kullanıcıdan gelen metin buraya gelecek."])

# SVM modeli ile tahmin yap
svm_tahmini = model_SVM.predict(kullanici_konu_vectorized)

# Random Forest modeli ile tahmin yap
rf_tahmini = model_RF.predict(kullanici_konu_vectorized)

# Test seti üzerinde SVM modelinin tahminlerini al
test_tahminleri_svm = model_SVM.predict(X_test)

# Test seti üzerinde Random Forest modelinin tahminlerini al
test_tahminleri_rf = model_RF.predict(X_test)

# Accuracy, F1 Score ve Classification Report değerlerini hesapla ve yazdır
accuracy_svm = accuracy_score(y_test, test_tahminleri_svm)
f1_svm = f1_score(y_test, test_tahminleri_svm, average='weighted')
classification_report_svm = classification_report(y_test, test_tahminleri_svm)

accuracy_rf = accuracy_score(y_test, test_tahminleri_rf)
f1_rf = f1_score(y_test, test_tahminleri_rf, average='weighted')
classification_report_rf = classification_report(y_test, test_tahminleri_rf)

# Sonuçları yazdır
print("SVM Modeli Sonuçları:")
print("Accuracy (SVM):", accuracy_svm)
print("F1 Score (SVM):", f1_svm)
print("Classification Report (SVM):\n", classification_report_svm)

print("\nRandom Forest Modeli Sonuçları:")
print("Accuracy (Random Forest):", accuracy_rf)
print("F1 Score (Random Forest):", f1_rf)
print("Classification Report (Random Forest):\n", classification_report_rf)
```
